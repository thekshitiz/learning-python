### Book Value
- Value of company if sold today
- Total assets - total liabilities

#### Lemonade Stand - Today
- Assets
    - Table + chair: 35 USD
    - Cups: 200 cups of 0.05 USD/cup
    - Lemonade mix: 25 USD
- Liabilities
    - None

```python
table_and_chair = 35
no_cups = 200
cup_price = 0.05
lemonade_mix = 25
```

```python
book_value = table_and_chair + no_cups * cup_price + lemonade_mix
book_value
```

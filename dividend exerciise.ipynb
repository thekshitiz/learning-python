**Earnings per share**

```python
earning = 600
shares_outstanding = 1000
```

```python
eps = earning / shares_outstanding
eps
```

**Price to earnings**

```python
share_price = 0.7
```

```python
pe_ratio = share_price / eps
pe_ratio
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.1, 5, 100)
```

```python
y = share_price / x
```

```python
plt.plot(x,y)
plt.xlabel("EPS")
plt.ylabel("P/E ratio")
```
